```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Parameters for our graph; we'll output images in a 4x4 configuration
nrows = 4
ncols = 4

# Index for iterating over images
pic_index = 0
```

```python
# Set up matplotlib fig, and size it to fit 4x4 pics
fig = plt.gcf()
fig.set_size_inches(ncols * 4, nrows * 4)

pic_index += 4
next_0_pix = [os.path.join(train_0_dir, fname) 
                for fname in train_0_names[pic_index-4:pic_index]]
next_5_pix = [os.path.join(train_5_dir, fname) 
                for fname in train_5_names[pic_index-4:pic_index]]
next_C_pix = [os.path.join(train_C_dir, fname) 
                for fname in train_C_names[pic_index-4:pic_index]]
next_Z_pix = [os.path.join(train_Z_dir, fname) 
                for fname in train_Z_names[pic_index-4:pic_index]]

for i, img_path in enumerate(next_0_pix+next_C_pix+next_Z_pix+next_5_pix):
  # Set up subplot; subplot indices start at 1
  sp = plt.subplot(nrows, ncols, i + 1)
  sp.axis('Off') # Don't show axes (or gridlines)

  img = mpimg.imread(img_path)
  plt.imshow(img)

plt.show()
```
